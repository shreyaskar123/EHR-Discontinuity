```python
import pandas as pd
medicare = pd.read_csv("/netapp2/home/se197/RPDR/Josh Lin/3_EHR_V2/CMS/Data/final_medicare.csv")
```

```python
def val_appended(card_outcomes): #MI, stroke, death
    for outcomes in card_outcomes:
        if (outcomes == 1):
            return 1
    return 0
out_comp_cardiovascular_R1 = [] 
for index in range(medicare.shape[0]):
        rel_outcomes = [medicare.at[index,'Out_IscheStroke_R1'], medicare.at[index,'Out_HemoStroke_R1'], medicare.at[index, 'Out_UncertainStroke_R1'], medicare.at[index, 'Out_MI_R1'], medicare.at[index,'ehr_death']]
        out_comp_cardiovascular_R1.append(val_appended(rel_outcomes))

medicare.insert(loc = 1, column = "Out_comp_cardiovascular_R1", value = out_comp_cardiovascular_R1)
```

```python
from datetime import date
import math
ehr_claims_death = []
for rows in medicare[['D_Index', 'D_Death_C']].itertuples():
    d0 = date(int(rows[1][0:4]), int(rows[1][5:7]), int(rows[1][8:10]))
    try: 
        math.isnan(float(rows[2]))
        ehr_claims_death.append(0)
        continue
    except ValueError:
       d1 = date(int(rows[2][0:4]), int(rows[2][5:7]), int(rows[2][8:10]))
       delta = d1 - d0
       print(delta.days)
       if (delta.days <= 365):
        ehr_claims_death.append(1)
        continue
       else:
        ehr_claims_death.append(0)
medicare['claims_death'] = ehr_claims_death 
  
```

```python
count = 0
t_val = 0
for values in ehr_claims_death: 
    t_val = t_val + 1
    if (values == 1): 
        count = count + 1
print(count, t_val)
```
